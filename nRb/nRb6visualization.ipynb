```python
import numpy
import matplotlib.pyplot as plt
data = numpy.loadtxt(fname='nRb6ex.csv', delimiter=';')
ax=data[:,0]
ay=data[:,1]
plot = plt.plot(ax,ay)
plt.xlabel('Wavelengths, nm')
plt.ylabel('Relative intensity, r.u.')
plt.savefig('nRb6_excitation.pdf', dpi=300)
plt.show()
```
